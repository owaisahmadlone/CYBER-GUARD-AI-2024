```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import BertModel, BertTokenizer
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

class MultiLabelDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len, label_cols):
        self.texts = list(dataframe['text'])
        self.labels1 = list(dataframe[label_cols[0]])
        self.labels2 = list(dataframe[label_cols[1]])
        
        self.tokenizer = tokenizer
        self.max_len = max_len
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        # Tokenize text
        encoding = self.tokenizer(
            self.texts[idx], 
            return_tensors='pt',
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            add_special_tokens=True
        )
        
        return {
            'input_ids': encoding['input_ids'].squeeze(0),  # Remove extra dimension
            'attention_mask': encoding['attention_mask'].squeeze(0),  # Remove extra dimension
            'labels1': torch.tensor(self.labels1[idx], dtype=torch.long),
            'labels2': torch.tensor(self.labels2[idx], dtype=torch.long)
        }
```

```python
def multi_task_loss(logits1, logits2, labels1, labels2):
    criterion = nn.CrossEntropyLoss()
    loss1 = criterion(logits1, labels1)
    loss2 = criterion(logits2, labels2)
    
    # You can add weights to balance the losses if needed
    total_loss = loss1 + loss2
    
    return total_loss
```

### Train the classifier layers only

```python
label_cols = ['label1', 'label2']
MAX_LEN=512
BERT_MODEL = 'bert-base-uncased'
```

```python
# from sklearn.model_selection import train_test_split
# train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
import torch
import torch.nn as nn
from transformers import BertModel, BertTokenizer

class BERTFrozenBaseClassifier(nn.Module):
    def __init__(self, bert_model_name, num_classes1, num_classes2):
        super(BERTFrozenBaseClassifier, self).__init__()
        
        # Load pretrained BERT model and freeze its parameters
        self.bert = BertModel.from_pretrained(bert_model_name)
        for param in self.bert.parameters():
            param.requires_grad = False
        
        # Get the hidden size of BERT model
        hidden_size = self.bert.config.hidden_size
        
        # Dropout for regularization
        self.dropout = nn.Dropout(0.1)
        
        # Two separate classification heads
        self.classifier1 = nn.Sequential(
            nn.Linear(hidden_size, hidden_size//2),
            nn.ReLU(),
            nn.Linear(hidden_size//2, num_classes1)
        )
        
        self.classifier2 = nn.Sequential(
            nn.Linear(hidden_size, hidden_size//2),
            nn.ReLU(),
            nn.Linear(hidden_size//2, num_classes2)
        )
    
    def forward(self, input_ids, attention_mask):
        # Get BERT outputs with no gradient computation
        with torch.no_grad():
            outputs = self.bert(
                input_ids=input_ids, 
                attention_mask=attention_mask
            )
        
        # Use the [CLS] token representation
        pooled_output = outputs.last_hidden_state[:, 0, :]
        
        # Apply dropout
        pooled_output = self.dropout(pooled_output)
        
        # Get logits for both classification tasks
        logits1 = self.classifier1(pooled_output)
        logits2 = self.classifier2(pooled_output)
        
        return logits1, logits2
```

```python
# Determine number of classes for each task
num_classes1 = len(categories)
num_classes2 = len(subcategories)

# Device setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Tokenizer and model
tokenizer = BertTokenizer.from_pretrained(BERT_MODEL)

# Create model
model = BERTFrozenBaseClassifier(
    bert_model_name=BERT_MODEL, 
    num_classes1=num_classes1, 
    num_classes2=num_classes2
).to(device)
```

```python
BATCH_SIZE=64
LR=5e-5

train_dataset = MultiLabelDataset(
        train_df, 
        tokenizer, 
        max_len=MAX_LEN, 
        label_cols=label_cols
    )
val_dataset = MultiLabelDataset(
    val_df, 
    tokenizer, 
    max_len=MAX_LEN, 
    label_cols=label_cols
)

# Data loaders
train_loader = DataLoader(
    train_dataset, 
    batch_size=BATCH_SIZE, 
    shuffle=True
)
val_loader = DataLoader(
    val_dataset, 
    batch_size=BATCH_SIZE
)

# Optimizer
optimizer = torch.optim.Adam(
    [
        {'params': model.classifier1.parameters()},
        {'params': model.classifier2.parameters()}
    ], 
    lr=LR
)

# Learning rate scheduler
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, 
    mode='min', 
    factor=0.5, 
    patience=2
)

# Loss function
criterion = nn.CrossEntropyLoss()
```

```python
# Training loop
best_val_loss = float('inf')

NUM_EPOCHS=5

for epoch in range(NUM_EPOCHS):
    # Training phase
    model.train()
    train_loss = 0
    
    for batch in train_loader:
        # Move to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels1 = batch['labels1'].to(device)
        labels2 = batch['labels2'].to(device)
        
        # Zero gradients
        optimizer.zero_grad()
        
        # Forward pass
        logits1, logits2 = model(input_ids, attention_mask)
        
        # Compute loss
        loss1 = criterion(logits1, labels1)
        loss2 = criterion(logits2, labels2)
        loss = loss1 + loss2
        
        # Backward pass
        loss.backward()
        
        # Optimize
        optimizer.step()
        
        train_loss += loss.item()
    
    # Validation phase
    model.eval()
    val_loss = 0
    correct1, correct2 = 0, 0
    total = 0
    
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels1 = batch['labels1'].to(device)
            labels2 = batch['labels2'].to(device)
            
            # Forward pass
            logits1, logits2 = model(input_ids, attention_mask)
            
            # Compute loss
            loss1 = criterion(logits1, labels1)
            loss2 = criterion(logits2, labels2)
            loss = loss1 + loss2
            
            val_loss += loss.item()
            
            # Compute accuracy
            _, pred1 = torch.max(logits1, 1)
            _, pred2 = torch.max(logits2, 1)
            total += labels1.size(0)
            correct1 += (pred1 == labels1).sum().item()
            correct2 += (pred2 == labels2).sum().item()
    
    # Average losses
    train_loss /= len(train_loader)
    val_loss /= len(val_loader)
    
    # Accuracy calculation
    acc1 = 100 * correct1 / total
    acc2 = 100 * correct2 / total
    
    # Print epoch summary
    print(f'Epoch {epoch+1}/{NUM_EPOCHS}:')
    print(f'Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')
    print(f'Accuracy Task 1: {acc1:.2f}%, Accuracy Task 2: {acc2:.2f}%')
    
    # Learning rate scheduling
    scheduler.step(val_loss)
    
    # Model checkpointing
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'best_frozen_bert_model.pth')
```
