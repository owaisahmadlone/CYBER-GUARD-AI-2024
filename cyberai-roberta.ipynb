```python
import torch

# Custom Dataset for Multi-Label Classification
class MultiLabelDataset(torch.utils.data.Dataset):
    def __init__(self, dataframe, tokenizer, max_len=512):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.max_len = max_len
    
    def __len__(self):
        return len(self.dataframe)
    
    def __getitem__(self, idx):
        text = str(self.dataframe.iloc[idx]['text'])
        category = self.dataframe.iloc[idx]['label1']
        subcategory = self.dataframe.iloc[idx]['label2']
        
        # Tokenize text
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'category_labels': torch.tensor(category, dtype=torch.long),
            'subcategory_labels': torch.tensor(subcategory, dtype=torch.long)
        }
```

```python
# Custom Model for Multi-Output Classification
class MultiOutputClassifier(torch.nn.Module):
    def __init__(self, model_name, num_categories, num_subcategories):
        super(MultiOutputClassifier, self).__init__()
        
        # Load base model
        self.base_model = AutoModelForSequenceClassification.from_pretrained(
            model_name, 
            num_labels=num_categories  # Initial loading with category labels
        )
        
        # Modify the classifier head for multi-output
        self.base_model.classifier = torch.nn.Sequential(
            torch.nn.Linear(self.base_model.classifier.in_features, 512),
            torch.nn.ReLU(),
            torch.nn.Dropout(0.3)
        )
        
        # Add separate classification heads
        self.category_classifier = torch.nn.Linear(512, num_categories)
        self.subcategory_classifier = torch.nn.Linear(512, num_subcategories)
    
    def forward(
        self, 
        input_ids=None, 
        attention_mask=None, 
        category_labels=None, 
        subcategory_labels=None
    ):
        # Get base model outputs
        outputs = self.base_model.base_model(
            input_ids=input_ids, 
            attention_mask=attention_mask
        )
        
        # Get pooled output
        pooled_output = outputs.last_hidden_state[:, 0, :]
        
        # Pass through shared layers
        shared_features = self.base_model.classifier(pooled_output)
        
        # Separate classification heads
        category_logits = self.category_classifier(shared_features)
        subcategory_logits = self.subcategory_classifier(shared_features)
        
        # Compute losses if labels are provided
        loss = None
        if category_labels is not None and subcategory_labels is not None:
            category_loss = torch.nn.CrossEntropyLoss()(category_logits, category_labels)
            subcategory_loss = torch.nn.CrossEntropyLoss()(subcategory_logits, subcategory_labels)
            loss = category_loss + subcategory_loss
        
        return {
            'loss': loss,
            'category_logits': category_logits,
            'subcategory_logits': subcategory_logits
        }
```

```python
# Custom Trainer to handle multi-output
class MultiOutputTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False):
        labels = {
            'category_labels': inputs.get('category_labels'),
            'subcategory_labels': inputs.get('subcategory_labels')
        }
        
        outputs = model(
            input_ids=inputs['input_ids'],
            attention_mask=inputs['attention_mask'],
            category_labels=labels['category_labels'],
            subcategory_labels=labels['subcategory_labels']
        )
        
        loss = outputs['loss']
        return (loss, outputs) if return_outputs else loss
```

```python
import numpy as np

def compute_metrics(pred):
    # Assuming predictions are structured as a 2D array with logits
    logits = pred.predictions  # This should be a single array of logits
    category_preds = np.argmax(logits[0], axis=-1)  # Adjust if the structure differs
    subcategory_preds = np.argmax(logits[1], axis=-1)  # Adjust based on the actual output

    # Accessing label_ids directly
    category_labels = pred.label_ids[0]  # Adjust if the structure differs
    subcategory_labels = pred.label_ids[1]  # Adjust based on the actual output

    # Calculate accuracies
    category_acc = np.mean(category_preds == category_labels)
    subcategory_acc = np.mean(subcategory_preds == subcategory_labels)

    return {
        'category_accuracy': category_acc,
        'subcategory_accuracy': subcategory_acc,
        'overall_accuracy': (category_acc + subcategory_acc) / 2
    }
```

```python
# Initialize tokenizer
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

# Create datasets
train_dataset = MultiLabelDataset(train_df, tokenizer)
val_dataset = MultiLabelDataset(val_df, tokenizer)

# Initialize multi-output model
model = MultiOutputClassifier(
    'distilbert-base-uncased', 
    num_categories=num_categories,
    num_subcategories=num_subcategories
)
```

```python
# Training Arguments
NUM_EPOCHS = 1
LR = 5e-4
EVAL_STEPS = 300  # Specify the number of steps after which to run evaluation
SAVE_STEPS = 300  # Change save_steps to be a multiple of eval_steps

training_args = TrainingArguments(
    output_dir='./multilabel_classification_results',
    num_train_epochs=NUM_EPOCHS,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    learning_rate=LR,
    evaluation_strategy="steps",  # Change to "steps"
    eval_steps=EVAL_STEPS,  # Specify the evaluation steps
    save_strategy="steps",
    save_steps=SAVE_STEPS,  # Set save_steps to a multiple of eval_steps
    load_best_model_at_end=True,
    metric_for_best_model='overall_accuracy',
    report_to=None,  # This disables WandB logging
    save_total_limit=1  # Limit the saved checkpoints to only the latest version
)
```

```python
# Initialize Trainer
trainer = MultiOutputTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

## Train the model
trainer.train()
```

```python
# Prediction Function
def predict_categories(text, model, tokenizer, category_encoder, subcategory_encoder):
    # Prepare input
    inputs = tokenizer.encode_plus(
        text,
        add_special_tokens=True,
        max_length=512,
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )
    
    # Predict
    model.eval()
    with torch.no_grad():
        outputs = model(
            input_ids=inputs['input_ids'],
            attention_mask=inputs['attention_mask']
        )
    
    # Get predictions
    category_pred = torch.argmax(outputs['category_logits'], dim=1).cpu().numpy()[0]
    subcategory_pred = torch.argmax(outputs['subcategory_logits'], dim=1).cpu().numpy()[0]
    
    # Decode predictions
    predicted_category = category_encoder.inverse_transform([category_pred])[0]
    predicted_subcategory = subcategory_encoder.inverse_transform([subcategory_pred])[0]
    
    return {
        'category': predicted_category,
        'sub_category': predicted_subcategory
    }
```
